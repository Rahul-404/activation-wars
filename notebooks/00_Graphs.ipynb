## Sigmoid function

```python

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate x values
x = np.linspace(-10, 10, 1000)
y = sigmoid(x)

# Plot
plt.figure(figsize=(4, 3))
plt.plot(x, y, 
        #  label='Sigmoid Function', 
         color='blue')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
# plt.legend()
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axhline(1, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.tight_layout()

# Save the plot as an image
plt.savefig('results/sigmoid_plot.png', dpi=100)

# Show the plot
plt.show()
```

## Derivative of sigmoid

```python

# Derivative of sigmoid: sigmoid(x) * (1 - sigmoid(x))
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

# Generate x values
x = np.linspace(-10, 10, 1000)
y = sigmoid(x)
y_deriv = sigmoid_derivative(x)

# Plot both sigmoid and its derivative
plt.figure(figsize=(5, 4))
plt.plot(x, y, label='Sigmoid', color='blue')
plt.plot(x, y_deriv, label='Sigmoid Derivative', color='red')
plt.title('Sigmoid Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.tight_layout()

# Save the plot
plt.savefig('results/sigmoid_and_derivative.png', dpi=100)

# Show the plot
plt.show()
```

## Sigmoid Non-Symmetry

```python
# Define sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate input values
x_vals = np.linspace(-10, 10, 400)
y_vals = sigmoid(x_vals)

# Plot sigmoid function
plt.figure(figsize=(5, 4))
plt.plot(x_vals, y_vals, label='Sigmoid', color='blue')
plt.axhline(0.5, color='gray', linestyle='--', linewidth=1, label='y = 0.5')
plt.axvline(0, color='black', linestyle='--', linewidth=1, label='x = 0')

# Highlight non-symmetry around zero
plt.plot([-2, 2], [sigmoid(-2), sigmoid(2)], 'ro', label='2 comparison')
plt.text(-2, sigmoid(-2) + 0.05, f'{sigmoid(-2):.2f}', ha='center', color='red')
plt.text(2, sigmoid(2) + 0.05, f'{sigmoid(2):.2f}', ha='center', color='red')

# Labels and title
plt.title("Sigmoid Function: Lack of Symmetry Around Zero")
plt.xlabel("Input (x)")
plt.ylabel("Sigmoid(x)")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save the plot
plt.savefig('results/sigmoid_and_symmetry.png', dpi=100)

# Show the plot
plt.show()
```

## Tanh function

```python
# Define tanh function
def tanh(x):
    return np.tanh(x)

# Generate input values
x = np.linspace(-10, 10, 1000)
y = tanh(x)

# Plot
plt.figure(figsize=(5, 4), dpi=100)
plt.plot(x, y, label='tanh(x)', color='darkorange')
plt.title('Tanh Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend()

# Save the figure
plt.savefig('results/tanh_plot.png', dpi=100, bbox_inches='tight')
plt.show()
```

## Tanh Derivative

```python
def tanh_derivative(x):
    return 1 - np.tanh(x) ** 2

# Generate input values
x = np.linspace(-10, 10, 1000)
y = tanh(x)
y_ = tanh_derivative(x)

# Plot
plt.figure(figsize=(5, 4), dpi=100)
plt.plot(x, y, label='tanh(x)', color='darkorange')
plt.plot(x, y_, label="tanh'(x)", color='teal')
plt.title('Derivative of Tanh Activation Function')
plt.xlabel('Input')
plt.ylabel("Derivative")
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend()

# Save the plot
plt.savefig('results/tanh_derivative_plot.png', dpi=100, bbox_inches='tight')
plt.show()
```

## ReLU function

```python
# Define ReLU
def relu(x):
    return np.maximum(0, x)

# Generate input values
x = np.linspace(-10, 10, 1000)
y = relu(x)

# Plot
plt.figure(figsize=(5, 4), dpi=100)
plt.plot(x, y, label='ReLU(x)', color='royalblue')
plt.title('ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend()


# Save the plot
plt.savefig('results/relu_plot.png', dpi=100, bbox_inches='tight')
plt.show()
```

## Plot "Dying ReLU" Effect

```python
# Define ReLU derivative
def relu_derivative(x):
    return np.where(x > 0, 1.0, 0.0)

# Input values
x = np.linspace(-10, 10, 1000)
y = relu(x)
y_ = relu_derivative(x)

# Plot
plt.figure(figsize=(5, 4), dpi=100)
plt.plot(x, y, label='ReLU(x)', color='royalblue')
plt.plot(x, y_, label="ReLU'(x)", color='green')
plt.plot(x, y_, label='Dead ReLU (always 0)', color='crimson', linestyle='--')
plt.title("Derivative of ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Derivative")
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend()

# Save plot
plt.savefig('results/relu_derivative_plot.png', dpi=100, bbox_inches='tight')
plt.show()
```

## Leaky ReLU Function

```python
# Leaky ReLU definition
def leaky_relu(x, alpha=0.1):
    return np.where(x > 0, x, alpha * x)

# Inputs and outputs
x = np.linspace(-10, 10, 1000)
y = leaky_relu(x)

alpha = 0.1

# Plot
plt.figure(figsize=(5, 4), dpi=100)
plt.plot(x, y, label=f'Leaky ReLU ( = {alpha})', color='darkviolet')
plt.title('Leaky ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)

# Annotations
plt.annotate(
    'Small positive slope\nin negative region',
    xy=(-6, leaky_relu(-6)), xytext=(-9, 3),
    arrowprops=dict(facecolor='black', arrowstyle='->'),
    fontsize=9
)

plt.annotate(
    'Positive slope\nin positive region',
    xy=(6, leaky_relu(6)), xytext=(2, 6),
    arrowprops=dict(facecolor='black', arrowstyle='->'),
    fontsize=9
)

plt.legend()

# Save plot
plt.savefig('results/leaky_relu_plot.png', dpi=100, bbox_inches='tight')
plt.show()
```

## Leaky ReLU Derivative

## Parametric ReLU Function

```python
# Parametric ReLU (PReLU) function
def prelu(x, alpha=0.1):
    return np.where(x >= 0, x, alpha * x)

# Generate input values
x = np.linspace(-10, 10, 1000)
alpha = 0.1
y = prelu(x, alpha)

# Plotting
fig, ax = plt.subplots(figsize=(6, 5), dpi=100)
ax.plot(x, y, label=f'PReLU (a = {alpha})', color='royalblue')
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.set_title("Parametric ReLU (PReLU) Activation Function")
ax.set_xlabel("Input")
ax.set_ylabel("Output")
ax.grid(True)
ax.legend()

# Save plot to the results folder
plt.savefig("results/prelu_activation_plot.png", format='png', bbox_inches='tight', dpi=100)
plt.show()
```

## Parametric ReLU Derivative

```python
# PReLU function and its derivative
def prelu(x, alpha=0.1):
    return np.where(x >= 0, x, alpha * x)

def prelu_derivative(x, alpha=0.1):
    return np.where(x >= 0, 1.0, alpha)

# Generate input values
x = np.linspace(-10, 10, 1000)
alpha = 0.1
y = prelu(x, alpha)
y_ = prelu_derivative(x, alpha)

# Plot
plt.figure(figsize=(5, 4), dpi=100)
plt.plot(x, y, label='PReLU(x) at a = 0.1', color='royalblue')
plt.plot(x, y_, label="PReLU'(x) at a = 0.1", color='crimson')
plt.title('Derivative of PReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output / Derivative')
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend()

# Save the plot
plt.savefig('results/prelu_derivative_plot.png', dpi=100, bbox_inches='tight')
plt.show()
```

## Exponential Linear Unit (ELUs) Function

```python
def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

# Generate input values
x = np.linspace(-5, 5, 1000)
alpha = 1.0
y = elu(x, alpha)

# Plot
plt.figure(figsize=(5, 4), dpi=100)
plt.plot(x, y, label='ELU(x)', color='mediumblue')
# plt.plot(x, y_, label="ELU'(x)", color='orangered')
plt.title('ELU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend()

# Save the plot
plt.savefig('results/elu_plot.png', dpi=100, bbox_inches='tight')
plt.show()
```

## Exponential Linear Unit (ELUs) Function Derivative

```python
def elu_derivative(x, alpha=1.0):
    return np.where(x > 0, 1.0, alpha * np.exp(x))  # or: elu(x, alpha) + alpha

# Generate input values
x = np.linspace(-5, 5, 1000)
alpha = 1.0
y = elu(x, alpha)
y_ = elu_derivative(x, alpha)

# Plot
plt.figure(figsize=(5, 4), dpi=100)
plt.plot(x, y, label='ELU(x) at a=1', color='mediumblue')
plt.plot(x, y_, label="ELU'(x) at a=1", color='orangered')
plt.title('Derivative of ELU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend()

# Save the plot
plt.savefig('results/elu_derivative_plot.png', dpi=100, bbox_inches='tight')
plt.show()
```

## Softmax Function

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))  # subtract max for numerical stability
    return e_x / np.sum(e_x)

# Input values for 3 example classes
x = np.linspace(-2, 2, 100)
softmax_outputs = np.stack([softmax([xv, 0, -xv]) for xv in x])

# Plot each class probability
plt.figure(figsize=(6, 4), dpi=100)
plt.plot(x, softmax_outputs[:, 0], label='softmax(x, 0, -x)', color='red')
plt.plot(x, softmax_outputs[:, 1], label='softmax(0)', color='green')
plt.plot(x, softmax_outputs[:, 2], label='softmax(-x)', color='blue')
plt.title('Softmax Outputs Over Input Range')
plt.xlabel('Input x')
plt.ylabel('Softmax Probability')
plt.legend()
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)

plt.savefig("results/softmax_plot.png", dpi=100, bbox_inches='tight')
plt.show()
```

## Swish Function

```python
# Swish activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def swish(x):
    return x * sigmoid(x)

# Generate input values
x = np.linspace(-10, 10, 1000)
y = swish(x)

# Plot
plt.figure(figsize=(5.5, 4), dpi=100)
plt.plot(x, y, label='Swish(x)', color='purple')
# plt.plot(x, y_, label="Swish'(x)", color='darkgreen')
plt.title('Swish Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend()

# Save the plot
plt.savefig('results/swish_plot.png', dpi=100, bbox_inches='tight')
plt.show()
```

## Swish Function Derivative

```python
def swish_derivative(x):
    sig = sigmoid(x)
    return sig + x * sig * (1 - sig)

# Generate input values
x = np.linspace(-10, 10, 1000)
y_ = swish_derivative(x)

# Plot
plt.figure(figsize=(5.5, 4), dpi=100)
plt.plot(x, y, label='Swish(x)', color='purple')
plt.plot(x, y_, label="Swish'(x)", color='darkgreen')
plt.title('Swish Activation Function and Its Derivative')
plt.xlabel('Input')
plt.ylabel('Output / Derivative')
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend()

# Save the plot
plt.savefig('results/swish_derivative_plot.png', dpi=100, bbox_inches='tight')
plt.show()
```
